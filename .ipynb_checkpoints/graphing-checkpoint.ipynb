```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
X = np.arange(-5, 5, 0.05)
```

```python
Y = np.vectorize(sign)(X)
```

```python
import math
```

```python
Y2 = np.vectorize(math.sin)(X / 2)
```

```python
%matplotlib inline
plt.plot(X, Y, "c")
plt.plot(X, Y2, "r")
```

```python
from mpl_toolkits import mplot3d
```

```python
%matplotlib inline
fig = plt.figure()
ax = plt.axes(projection='3d')
XR = np.random.rand(10)
YR = np.random.rand(10)
ZR = np.random.rand(10)
ax.scatter3D(XR, YR, ZR, c=YR, cmap="Greens")
```

```python
%matplotlib inline
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.view_init(math.asin(math.tan(30 * (math.pi / 180))) * (180 / math.pi), 45)

def f(x, y):
    return np.sin(x / 5) * (np.sin(y) / 10)

x = np.linspace(-6, 6, 300)
y = np.linspace(-6, 6, 300)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

ax.contour3D(X, Y, Z, 250)
```
